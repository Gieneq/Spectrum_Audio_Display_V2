# Sine wave generation for DAC

128 samples amplitude 1.

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
from math import pi

plots_scale = 0.8
plt.rcParams['figure.figsize'] = [plots_scale*15, plots_scale*12]

# Twice as many samples are needed
samples_count = 128

frequency = 1.0/samples_count

print(f"Generating {samples_count} samples, frequency {round(frequency, 4)}Hz")

offsetVal = 128
minVal =    offsetVal
maxVal =    2**12-1 - offsetVal

sample = np.arange(samples_count)

# val = np.sin(2*pi*frequency*sample)
val = np.round((maxVal-minVal)*(np.sin(2*pi*frequency*sample)+1)/2 + minVal).astype('int')

print(f"Min: {min(val)}/{minVal}, max: {max(val)}/{maxVal}.")

plt.plot(sample[:len(sample)], val[:len(sample)], '*r')
plt.xticks(fontsize= 18)
plt.yticks(fontsize= 18)
plt.grid(color='lightgray', linestyle='-', linewidth=1)
plt.xlabel("Sample[1]", fontsize=18)
plt.ylabel(r'Amplitude A[1]', fontsize=18)
plt.title(f"Generated {samples_count} sine samples", fontsize=18)
plt.show()

str_samples = str([v for v in val.tolist()])[1:-1]

print(f"#define DAC_SINE_SAMPLES_COUNT {samples_count}")
print(f"const uint16_t DAC_SINE_SAMPLES[{samples_count}] = " + r"{" + str_samples + r"};")


```
